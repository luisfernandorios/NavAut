# Uso de Datasets en Scikit Learn

# Se importan los datasets

```python
#se usa como ejemplo la clasificación de flores Iris desarrollado por Ronald Fisher
from sklearn import datasets
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

# Se imprimen las variables independientes, o features, del dataset 

```python
print(iris.data) 
```

```python
print(iris.feature_names) 
```

# Ahora se muestra la variable dependiente, etiqueta o label

```python
print(iris.target)   
print(iris.target_names)  
```

# Generando datos aleatorios para verificación de modelos

### Se generan datos relacionados linealmente
### https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

#se generan 100 datos para una sola feature con una desviación estándar del ruido Gaussiano agregado
X, y = make_regression(n_samples=100, n_features=1, noise=5.4)
plt.scatter(X,y)
```

### Dataset adecuado para Clustering
### https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

#500 muestras organizadas en tres blobs
X, y = make_blobs(500, centers=3)  
print(X.shape)
print(y.shape)
```

```python
# se grafican los tres blobs con los colores indicados                                   
rgb = np.array(['r', 'g', 'b'])
plt.scatter(X[:, 0], X[:, 1], color=rgb[y])
```

### Se crea un dataset organizado en forma circular
### https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles
#se crean un par de círculos concéntricos con radios distintos
X, y = make_circles(n_samples=100, noise=0.09)
print(X.shape)
print(y.shape)
```

```python
rg = np.array(['r', 'g'])
plt.scatter(X[:, 0], X[:, 1], color=rg[y])
```
